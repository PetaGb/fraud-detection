```python
import openml
import pandas as pd
```

```python
openml.datasets.list_datasets(output_format="dataframe")

# Get dataset by ID
# https://www.openml.org/search?type=data&status=active&id=46359&sort=runs
dataset = openml.datasets.get_dataset(46359)

# Get the data itself as a dataframe (or otherwise)
X, y, _, _ = dataset.get_data(dataset_format="dataframe")
```

```python
print(X.shape)
print()
X.head()
```

```python
if y is not None:
    df = X.copy()
    df['target'] = y
else:
    df = X
```

```python
print(df.shape)
print()
df.head()
```

```python
df.to_csv('fraud_df.csv',index=False)
```
